We know from last class that a Stack is an ordered collection of elements where items are added and removed from the top.

*Queues* are ordered collections of elements where items are added to the back and removed (andf returned) from the front.


```python
# From scratch implementation of Stack

class Stack:
    class __Node:
        def __int__(self, datum):
            self.datum = datum
            self.below = None

    def __int__(self):
        self.top = None
        self.count = 0

    def push (self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.top:
            self.top = new_node
        else:
            new_node.below = self.top
            self.top = new_node

    def pop(self):
        if self.top:
            self.count -= 1
            backup = self.top.datum
            self.top = self.top.below
            return backup
        raise IndexError("Stack is empty")

    # Nice3 to have methods
    def peek(self):
        if self.top:
            return self.top.datum
        raise IndexError("Stack is empty")
            
    def size(self):
        #count = 0
        #current = self.top
        #while current:
            #count += 1               # while current !=none 
            #current = current.below
        #return count
        return self.size

    def is_empty(self):
        return self.top == None 

        
```
