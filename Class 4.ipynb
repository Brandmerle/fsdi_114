````python
class DoublyLinkedList:
    class __Node:
        def __init__(self, datum):
            self.datum = datum
            self.next = None
            self.prev = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        if index < 0 or index > self.count:
            raise IndexError("Index not found")
        self.count += 1
        if index == 0:
            new_node.next = self.head
            if self.head:
                self.head.prev = new_node
            self.head = new_node
            if not self.tail:
                self.tail = new_node
        elif index == self.count - 1:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
        else:
            current = self.head
            for _ in range(index):
                current = current.next
            previous = current.prev
            new_node.next = current
            new_node.prev = previous
            current.prev = new_node
            previous.next = new_node

    def remove(self, value):
        current = self.head
        found = False
        while current and not found:
            if current.datum == value:
                found = True
            else:
                current = current.next
        if found:
            self.count -= 1
            if not current.prev:  # Removing the head
                self.head = current.next
                if self.head:
                    self.head.prev = None
                if current == self.tail:
                    self.tail = None
            elif not current.next:  # Removing the tail
                current.prev.next = None
                self.tail = current.prev
            else:  # Removing from the middle
                current.prev.next = current.next
                current.next.prev = current.prev
        else:
            raise ValueError("Target value not found")

    def index(self, value):
        count = 0
        current = self.head
        while current:
            if current.datum == value:
                return count
            current = current.next
            count += 1
        raise ValueError("Value not Found")

    def __str__(self):
        out = "["
        current = self.head
        while current:
            out += str(current.datum)
            current = current.next
            if current:
                out += ", "
        out += "]"
        return out

    def __len__(self):
        return self.count

    def remove_at(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of bounds")
        self.count -= 1
        if index == 0:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
            if not self.head:
                self.tail = None
        elif index == self.count: # index was the last element
            self.tail = self.tail.prev
            self.tail.next = None
        else:
            current = self.head
            for _ in range(index):
                current = current.next
            current.prev.next = current.next
            current.next.prev = current.prev

    def get(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of bounds")
        current = self.head
        for _ in range(index):
            current = current.next
        return current.datum

    def clear(self):
        self.head = None
        self.tail = None
        self.count = 0

    def reverse(self):
        current = self.head
        self.head, self.tail = self.tail, self.head
        while current:
            next_node = current.next
            current.next = current.prev
            current.prev = next_node
            current = next_node
````
