```python
# From scratch implementation of SinglyLinkedList

class SinglyLinkedList:
    class __Node:
        def __init__(self, datum):
            self.datum = datum
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.head:                #if self.head = none
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node  #updates current node next datum in current tail to point ot the new node
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        if index < 0 or index > self.count:
            raise IndexError("Index not found")
        self.count += 1
        if index == 0:
            new_node.next = self.head
            self.head = new_node
            if self.tail is None: 
                self.tail = new_node
        else:
            current = self.head
            previous = None
            for _ in range(index):
                previous = current
                current = current.next
            new_node.next = current
            previous.next = new_node
            if current is None: 
                self.tail = new_node
      

    def remove (self, value):
        #this mehtod removes the first instance of "value" and raises a ValueError if the value is not present on the lsit
        current = self.head
        found = False
        prev = None
        while current.next and not found:
            if current.datum == value:
                found = True
            else:
                prev = current
                current = current.next
        if found:
            self.count -= 1
            if not prev:
                # This means the target value is at the head of our list
                self.head = self.head.next
            else:
                # This means the target value is somewhere in the middle or the tail of the list
                prev.next = current.next
                if current == self.tail:
                    self.tail == prev
            if not self.head:
                self.tail = None
        else:
            raise ValueError("Target value not found")

    def index (self, value):
        #This method returns the positon in the list of the given "value" if it is present in the list
        #Otherwise, it raises a ValueError if the value is not present in the list
        count = 0
        current = self.head
        found = False
        while current and not found:
            if current.datum == value:
                found = True
            else:
                count += 1
                current = current.next
        if found:
            return current
        raise ValueError ("Value not Found")

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.datum
            current = current.next
            while current:                         # while current is not equal to None
                out += ", %s" % current.datum
                current = current.next
            out += "]"
            return out

    def __len__(self):
        #this method should return the total number of nodes in this list
        return self.count
```
